```python
## importanto as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o arquivo em excel para o dataframe df, negando as três primeiras linhas e as três últimas linhas do arquivo

df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx',skiprows=3, skipfooter = 3) 

df.head(3)
```

```python
df.tail()
```

```python
#remove os espaços em branco do nome das colunas
df.columns = df.columns.str.strip()
df.columns
```

```python
#apresenta a informação do dataframe
df.info()
```

```python
# apresenta alguns dados estatísticos das variávies númericas (Inteiros e reais)
print(df['Valor Tarifa Comercial'].describe()) 
print(df['Valor Tarifa Governo'].describe()) 
print(df['Valor Tarifa Embarque'].describe()) 
print(df['Valor Bilhete'].describe()) 
```

```python
# apresenta os dez maiores valores do campo "valor Tarifa Comercial"
df.nlargest(10,'Valor Bilhete')
```

```python
#apresenta somente os valores que estão acima do valor médio da tarifa comercial
df[df['Valor Tarifa Comercial']> df['Valor Tarifa Comercial'].mean()]['Companhia Aérea'].value_counts().plot.barh()
```

```python
#Apresenta a companhia com maior número de registros
df['Companhia Aérea'].value_counts().plot.barh()
```

```python
#Retorna o número de companhias aéreas que cada orgão utilizou. 
df.groupby('Nome do órgão superior')[['Companhia Aérea']].nunique()
```

```python
#apresenta o em forma de gráfico as
df[df['Cancelado'].str.strip()=='Sim']['Companhia Aérea'].value_counts().plot.barh()
```
